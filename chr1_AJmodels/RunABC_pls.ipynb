## 3. Estimate parameters

```python
%%bash
PLS=10
echo "task estimate
simName input_ABCtoolbox_M2_HPC_OSG_2.transformed_${PLS}pls.txt
obsName real_output_M23_IBD_transformed_${PLS}pls.summary
params 10-15,18,20-23
numRetained 1000
maxReadSims 1446124
diracPeakWidth 0.01
posteriorDensityPoints 100
jointPosteriors Log10_NWA,Log10_NEA
jointPosteriorDensityPoints 100
writeRetained 0
outputPrefix ABC_M2_genome_estimate_1446124_${PLS}pls_1000ret_
logFile ABC_M2_genome_estimate_1446124_${PLS}pls_1000ret.log
verbose" >test_ABC_estimate_PLS.txt
```
