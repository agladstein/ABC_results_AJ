## 2. Transform stats

ABCtoolbox input file, `test_ABC_transform_real.txt`

````python
%%bash
cd /vol_c/ABC_AJmodels_genome/PLS/dropIBDq
echo "task transform
input results_real_output_M23_dropIBDq.txt
output results_real_output_M23_dropIBDq_transformed.txt
linearComb Routput_input_ABC_HPC_OSG_CHTC_2_dropIBDq.txt
boxcox 1
logFile results_real_output_M23_dropIBDq_transformed.log
verbose
" >test_ABC_transform_sim.txt
````

run ABCtoolbox to transform stats

````python
%%bash
cd /vol_c/ABC_AJmodels_genome/PLS/dropIBDq
ABCtoolbox test_ABC_transform_sim.txt
````

## 3. Estimate parameters

### 3.a. Reduce number of components to use for estimation

````python
%%bash
cd /vol_c/ABC_AJmodels_genome/PLS/dropIBDq
cut -f-24,25-34 input_ABC_HPC_OSG_CHTC_2_dropIBDq_transformed.txt >input_ABC_HPC_OSG_CHTC_2_dropIBDq_transformed_10pls.txt
cut -f2-11 results_real_output_M23_dropIBDq_transformed.txt >results_real_output_M23_dropIBDq_transformed_10pls.txt
````

````python
%%bash
cd /vol_c/ABC_AJmodels_genome/PLS/dropIBDq
cut -f-24,25-44 input_ABC_HPC_OSG_CHTC_2_dropIBDq_transformed.txt >input_ABC_HPC_OSG_CHTC_2_dropIBDq_transformed_20pls.txt
cut -f2-21 results_real_output_M23_dropIBDq_transformed.txt >results_real_output_M23_dropIBDq_transformed_20pls.txt
````

### 3.b Estimate with ABCtoolbox

````python
%%bash
cd /vol_c/ABC_AJmodels_genome/PLS/dropIBDq
echo "task estimate
simName input_ABC_HPC_OSG_CHTC_2_dropIBDq_transformed_10pls.txt
obsName results_real_output_M23_dropIBDq_transformed_10pls.summary
params 2-6,8-15,17-23
numRetained 1000
maxReadSims 96334
diracPeakWidth 0.01
posteriorDensityPoints 100
jointPosteriors Log10_NWA,Log10_NEA
jointPosteriorDensityPoints 100
writeRetained 0
outputPrefix ABC_M2_genome_estimate_96334_10pls_1000ret_
logFile ABC_M2_genome_estimate_96334_10pls_1000ret.log
verbose" >test_ABC_estimate_PLS.txt
````

````python
%%bash
cd /vol_c/ABC_AJmodels_genome/PLS/dropIBDq
ABCtoolbox test_ABC_estimate_PLS.txt
````
